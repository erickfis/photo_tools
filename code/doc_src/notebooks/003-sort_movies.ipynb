# Sort movies

```python
from pathlib import Path
from tqdm import tqdm
import shutil
```

```python
source = r'C:\Users\erickfis\Videos\videos-onedrive'
dest = r'C:\Users\erickfis\Documents\onedrive_hotmail\OneDrive\video'
```

```python
all_files = list(Path(source).rglob('*.mp4'))
all_files
```

```python
len(all_files)
```

```python
file = all_files[-1]
for file in tqdm(all_files):
    date_info = file.stem.split('Vid')[1][:8]
    year = file.stem.split('Vid')[1][:4]
    month = file.stem.split('Vid')[1][4:6]
    new_folder = Path(dest) / year / month 
    new_path = new_folder / file.name
    new_folder.mkdir(parents=True, exist_ok=True)
    shutil.copy(file, new_path)
```
