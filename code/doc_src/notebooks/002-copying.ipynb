# Copying photos

```python
from pathlib import Path
from tqdm import tqdm
import shutil
```

```python
source = r'I:\fotos\erick'
dest = r'C:\Users\erick\OneDrive\Pictures\Camera Roll'
```

```python
all_pics = Path(source).rglob('*.JPG')
all_pics = [item for item in all_pics if item.parts[3].startswith('20')]
all_pics
```

```python
len(all_pics)
```

```python
file = all_pics[-1]
file
```

```python
'/'.join(file.parent.parts[-2:])

```

```python
errors = []
for file in tqdm(all_pics):
    try:
        parent = '/'.join(file.parent.parts[-2:])
        dest_folder = Path(dest) / parent
        dest_folder.mkdir(exist_ok=True, parents=True)
        new_path = dest_folder / file.name
        if new_path.exists:
            continue
        else:
            shutil.copy(file, new_path)
    except:
        errors.append(file)
```
